```python
def search_name(school_name):
    driver.get('https://dlt.aals.org/dlts')
#     elem = driver.find_element_by_xpath("/html/body/div/div/div/form/div[1]/div[1]/div/div/a/span")
#     elem.click()
    driver.find_element_by_xpath('//*[@id="q_contact_info_last_name_cont"]').send_keys(school_name)
    driver.find_element_by_xpath("/html/body/div/div/div/form/div[2]/div/input").click()
    try:
        example = driver.find_element_by_xpath('/html/body/div/div/div/table/tbody/tr/td/p')
        if example.text == 'No entries matching your search criteria were found.':
            name_exists = False
    except:
        name_exists = True
#         driver.find_element_by_xpath('//*[@id="q_tenured_and_dean"]').click()
#         driver.find_element_by_xpath("/html/body/div/div/div/form/div[2]/div/input").click()
    return name_exists
```

```python
def get_page_data(html, cur_data):
    dfs = pd.read_html(html)
    data = dfs[0].iloc[:,:-1]
    cur_data = pd.concat([cur_data, data])
    return cur_data
def create_path(*args):
    cur_path = os.getcwd()
    for value in args:
        cur_path  = os.path.join(cur_path, value)
    return cur_path
```

```python
min_law_df = pd.read_excel('Minority Law Teachers.xlsx', header = None)
min_law_df.columns = ['First Name', 'Last Name', 'School']
min_names = [x[0] + ' ' + x[1] for x in zip(min_law_df['First Name'], min_law_df['Last Name'])]
min_last_names = min_law_df['Last Name']
min_names_schools = list(zip(min_names, min_law_df['School']))
min_last_names_schools = list(zip(min_last_names, min_law_df['School']))
```

```python
cur_data = pd.DataFrame() 
aals_name_list = []
aals_school_list = []
fuzz_list = []
for i in range(len(min_names)):
    name_exists = search_name(min_last_names[i])
    max_fuzz_ratio = 0
    if name_exists:   
        html = driver.page_source
        cur_data = get_page_data(html, cur_data)
        soup = bs.BeautifulSoup(html, 'lxml')
        next_page_url = soup.find('a', {'rel': 'next'})
        while next_page_url:
            next_link = next_page_url.get('href')
            link = 'https://dlt.aals.org' + next_link
            driver.get(link)
            html = driver.page_source
            cur_data = get_page_data(html, cur_data)
            soup = bs.BeautifulSoup(html, 'lxml')
            next_page_url = soup.find('a', {'rel': 'next'})
        best_name = 'NA'
        best_school = 'NA'
        cur_data.reset_index(inplace = True, drop = True)
        for j in range(len(cur_data)):
            aals_name = cur_data['First Name'][j] + ' ' + cur_data['Last Name'][j]
            fuzz_ratio = fuzz.ratio(min_names[i], aals_name)
            if fuzz_ratio > max_fuzz_ratio:
                max_fuzz_ratio = fuzz_ratio
                best_name = aals_name
                best_name_school = cur_data['School'][j]
        aals_name_list.append(best_name)
        aals_school_list.append(best_name_school)
        fuzz_list.append(max_fuzz_ratio)
    else:
        print('{} is not in the database'.format(min_names[i]))
        best_name = 'NA'
        best_school = 'NA'
        aals_name_list.append(best_name)
        aals_school_list.append(best_school)
        fuzz_list.append(max_fuzz_ratio)
    
cur_data.to_csv(create_path('school_data', 'full_data_new.csv'))
```

```python
min_law_df['AALS Name'] = aals_name_list
min_law_df['AALS School'] = aals_school_list
min_law_df['Fuzz List'] = fuzz_list
```
