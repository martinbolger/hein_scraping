```python
import pandas as pd
import numpy as np
import itertools
```

```python
edge_df = pd.read_csv('edgelist.csv', header=None)
edge_df.dropna(how='all')
synons = pd.read_csv('synons.csv')
```

```python
comma_df = synons[['old', 'new']].dropna(how='all')
syn_df = synons[['syn1', 'syn2', 'syn3', 'syn4', 'syn5']].dropna(how='all')
```

```python
comma_df.insert(column  = 'old1', value = pd.Series([x.split(',')[0] for x in comma_df['old']]), loc = 1)
comma_df.insert(column  = 'old2', value = pd.Series([x.split(',')[1] for x in comma_df['old']]), loc = 2)
```

```python
def find_index_str(df, term):
    index_lists = [df.index[df[col] == term] for col in df.columns]
    index_list = [(index, i) for i, sublist in enumerate(index_lists) for index in sublist]
    return index_list
    
for i in range(len(comma_df)):
    first_term = comma_df['old1'][i].strip()
    second_term = comma_df['old2'][i].strip()
    update_term = comma_df['new'][i].strip()
    old = comma_df['old'][i].strip()
    num_commas = old.count(',')
    index_list = find_index_str(edge_df, first_term)
    for index in index_list:
        #Get the row of the dataframe and remove na values
        cur_series = edge_df.iloc[index[0], :].dropna()
        print(cur_series)
        #Check that the next term in the dataframe is the term we are expecting
        #If it is, update the previous term
        print(second_term)
        if cur_series[index[1] + 1] == second_term:
            
            edge_df.loc[index[0], [index[1]]] = update_term
            
            edge_df.loc[index[0], list(range(index[1] + 1, len(cur_series)-num_commas))] = list(cur_series[index[1] + 1 + num_commas:])

            edge_df.loc[index[0], list(range(len(cur_series)-num_commas, len(cur_series)))] = num_commas*[np.nan]
```

```python
for i in range(len(comma_df)):
    target_word = syn_df['syn1'][i]
    syns = list(syn_df.loc[i, ['syn1', 'syn2', 'syn3', 'syn4', 'syn5']].dropna())
    print(target_word)
    print(syns)
    index_list = []
    for word in syns:
        index_list = index_list + find_index_str(edge_df, word)
    index_list = sorted(index_list, key=lambda x: x[0], reverse = False)
    print(index_list)
    cur_row = -1
    for index in index_list:
        if index[0] != cur_row:
            cur_row = index[0]
            edge_df.loc[index[0], [index[1]]] = target_word
        else:
            print(index)
            topic_series = list(edge_df.iloc[index[0],:].dropna())
            print(topic_series[index[1] + 1:])
            edge_df.loc[index[0], list(range(index[1], len(topic_series)-1))] = topic_series[index[1] + 1:]
            print(edge_df.loc[index[0], len(cur_series)-1])
            edge_df.loc[index[0], len(topic_series)-1] = np.nan
```

```python
edge_df.to_csv('edgelist_update.csv')
```
