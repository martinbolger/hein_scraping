```python
import pandas as pd
import os

def create_path(*args):
    cur_path = os.getcwd()
    for value in args:
        cur_path  = os.path.join(cur_path, value)
    return cur_path
```

```python
files = os.listdir(create_path('author_papers', 'multi school data'))
```

```python
df = pd.DataFrame(columns = ['Title', 'List Author', 'Author(s)', 'Journal', 'BBCite', 'Cited (articles)', 'Cited (cases)', 'Accessed', 'Topics'])
for file in files:
    author_name = file.split('_')[0]
    new_data = pd.read_csv(create_path('author_papers', 'multi school data', file))
    if "List Author" not in new_data.columns:
        new_data.insert(1, "List Author", [author_name]*len(new_data)) 
    df = df.append(new_data, sort = False)
```

```python
def get_journal_data(jname_series):
    df = pd.DataFrame(columns = ['Journal', 'Vol', 'Issue', 'Pages'])
    for jname_string in jname_series:
        data_stream = dict.fromkeys(['Journal', 'Vol', 'Issue', 'Pages'], '')
        if jname_string != '':
            data_stream['Journal'] = jname_string.split(', Vol.', 1)[0]
            jelements = 'Vol. ' + jname_string.split('Vol.', 1)[1]
            jelement_list = jelements.split(',')

            for elm in jelement_list:
                if 'Vol' in elm:
                    data_stream['Vol'] = elm.strip()
                elif 'Issue' in elm:
                    data_stream['Issue'] = elm.split('Issue')[1].strip()
                elif 'pp.' in elm:
                    data_stream['Pages'] = elm.split('pp.')[1].strip()
        df = df.append(pd.DataFrame([list(data_stream.values())], columns = ['Journal', 'Vol', 'Issue', 'Pages']), sort=False)
    return df

df_j = get_journal_data(df['Journal'])
df_j
```
